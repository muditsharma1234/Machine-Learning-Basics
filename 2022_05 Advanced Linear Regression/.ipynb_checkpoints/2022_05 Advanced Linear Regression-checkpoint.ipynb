Linear Regression model is given by <center> $Y = \beta_0 + \beta_1X_1 + \beta_2X_2 + ... + \beta_pX_p + \epsilon$ </center>

- Prediction Accuracy: When the number of observations is a lot larger than the number of variables the least-square methods tend to perform better on the test observations. However, when the number of variables is a lot higher the least-square method may have very high variance and therefore leading to OVERFITTING. Overfitting causes poor performance on the test datasets. We can use some of the advanced linear regression models to remove variance significantly at cost of small increase in bias. 

- Model Interpretability: Some of the advanced linear regression techniques can do an automatic feature selection so as to remove unecessory variables. 

There can be multiple approaches to improve Prediction Accuracy or Model Interpretability. To list a few:
1. Subset Selection: Where we try to reduce the number of variables iteratively, wherein we start approaching situation where we are left with the most important predictor variables. This we did in our simple linear regression exercise via **BACKWARD ELIMINATION**
2. Shrinkage: Where using regularization models we tend to reduce the values of coefficients. Depending on type of regularization techniques some of the coefficients may even end up becoming Zero, in others it might become very close to Zero. 
3. Dimension Reduction: This approach involves projecting p-predictors to M-dimensional sub-space where M < p and subsequently these M subspaces are used for prediction using the least-square method. 

#### Algorithm for Subset Selection
For subset selection we select number of variables to be used for prediction and then select the best models out of possible models and then subsequently find out the best model.

The algorithm works in following steps:
- $M_0$ where the base model is selected with no predictors
- for x = 1,2, .... n where n is the total number of predictors in the base data
    - Fit all $(\frac{n}{x})$ models and choose the best model for appropriate x
    - Subsequently amongst all x's choose the best model 

Forward Selection: Addition of a variable one-by-one starting with a certain number of variables.

However, one question remains why does the regularization lead to a better model than the traditional regression? Let us start investigating this mathematically. 

```python
def compute_MSE(model, y_train, X_train):
    ypred = model.predict(X_train)
    return rmse(ypred,y_train)*len(y_train)
```

```python
compute_MSE(lasso,y_train,X_train)
```

Let us try to plot variance 

```python
MSE = []
y = np.arange(0.01,1,0.001)
for count,i in enumerate (y):
    lasso = Lasso(alpha = i, max_iter = 500)
    lasso.fit(X_train,y_train)
    MSE.append(compute_MSE(lasso,y_train,X_train))
```

```python
plt.plot(y,MSE)
```
